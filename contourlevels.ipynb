```python
import ipywidgets as widgets
import pyhf
import json
import matplotlib.pyplot as plt
import numpy as np
from pyhf.contrib.viz import brazil
import matplotlib
import matplotlib.pyplot as plt
```

```python
# create data
x = np.random.normal(size=1000)
y = np.random.normal(size=1000)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10.5, 9)
plt.hist2d(x,y, bins=(100, 100), cmap=plt.cm.jet)
plt.show()
```

```python
counts,ybins,xbins,image = plt.hist2d(x,y,bins=10)
```

```python
plt.contour(counts,extent=[xbins.min(),xbins.max(),ybins.min(),ybins.max()],levels=np.array([5.0,15.0,29.5]))
```

```python
plt.contour(counts,extent=[xbins.min(),xbins.max(),ybins.min(),ybins.max()],levels=np.array([5.0,15.0,29.5]),cmap='Greys')
plt.contourf(counts,extent=[xbins.min(),xbins.max(),ybins.min(),ybins.max()],levels=np.array([5.0,15.0,29.5]), cmap='RdGy')
```

```python
@widgets.interact(bins=(10, 100, 10))
def interact(bins=10):
        fig, (ax1,ax2,ax3) = plt.subplots(1,3)
        fig.set_size_inches(18, 6)
        counts,ybins,xbins,image = ax1.hist2d(x,y,bins)
        
        ax2.contour(counts,extent=[xbins.min(),xbins.max(),ybins.min(),ybins.max()],levels=np.array([5.0,15.0,29.5]))
        
        ax3.contourf(counts,extent=[xbins.min(),xbins.max(),ybins.min(),ybins.max()],levels=np.array([5.0,15.0,29.5]), cmap='RdGy')
       
```
